```python
#Task 2
class Assassin:
  total = 0

  def __init__(self,name,successrate):
    self.name=name
    self.sr=successrate
    Assassin.total+=1
  
  @classmethod
  def failureRate(cls,name,fr):
    cls.name=name
    cls.succrate=100 - fr
    return Assassin(cls.name,cls.succrate)
  @classmethod
  def failurePercentage(cls,name,fp):
    cls.name=name
    cls.succrate=100 - int(fp[0:2])
    return Assassin(cls.name,cls.succrate)

  def printDetails(self):
    print(f'Name: {self.name}\nSuccess rate: {self.sr}\nTotal number of Assasssin:{Assassin.total}')


john_wick = Assassin('John Wick',100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa",20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()

```

```python
#TASK 8
class Cylinder:
  radius = 5
  height = 18
  def __init__(self,rad,hgt):
    
    print(f"Default: radius={Cylinder.radius} height={Cylinder.height}")
    print(f"Updated: radius={rad} height={hgt}")
    self.radius = rad
    self.height = hgt
    Cylinder.radius = rad
    Cylinder.height = hgt

  @staticmethod
  def area(rad,hgt):
    x = 2*(3.1416)*rad**2 + 2*(3.1416)*rad*hgt
    print("Area:",x)

  @staticmethod
  def volume(rad,hgt):
    x = (3.1416)*rad**2*hgt
    print("Volume:",x)

  @classmethod
  def changeFormat(cls,s1):
    x,y = s1.split("-")
    obj = Cylinder(int(x),int(y))
    return obj

  @classmethod
  def swap(cls,rad,hgt):
    obj = Cylinder(hgt,rad)
    return obj

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#TASK10
class SultansDine:
  total_branch=0
  total_sell=0
  branch_name={}
  def __init__(self,branch):
      self.branch=branch
      SultansDine.total_branch=SultansDine.total_branch+1
    
  def sellQuantity (self, qty):
      self.qty=qty
      if self.qty<10:
          self.branch_sell=self.qty*300
      elif self.qty<20 and self.qty>=10:
          self.branch_sell=self.qty*350
      else:
          self.branch_sell=self.qty*400

      SultansDine.total_sell=SultansDine.total_sell + self.branch_sell
      SultansDine.branch_name[self.branch]=self.branch_sell

  @classmethod
  def details(cls):
      if cls.total_branch == 0:
          print(f'Total Number of branch(s): {cls.total_branch}')
          print(f'Total Sell: {cls.total_sell} Taka')
      else:
          print(f'Total Number of branch(s): {cls.total_branch}')
          print(f'Total Sell: {cls.total_sell} Taka')

          for branches, sell in cls.branch_name.items():
              sell_percentage=(sell/cls.total_sell)*100
              sell_percentage=round(sell_percentage,2)
              print(f'Branch name: {branches}, Branch Sell: {sell} Taka, Branch consists of total sells: {sell_percentage}%')

  def branchInformation (self):
      print(f'Branch Name:{self.branch}')
      print(f'Branch Sell={self.branch_sell}')

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
