```python
#Task 13
class Account:
  def __init__(self,name=None,balance=None):

    self.name=name
    self.balance=balance
    if self.name==None and self.balance==None:
      self.name='Default Name'
      self.balance=0.0
    elif self.name!=None and self.balance==None:
      self.balance=0.0
    else:
      self.balance=balance
  
  def details(self):
    return(f'{self.name}\n{self.balance}')
  
  def withdraw(self,amn):
    self.tak=None
    self.toti=amn
    self.tak=self.balance-float(self.toti)
    if self.balance==0.0 or self.balance<self.toti or self.tak==3070.0:
      print('Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.')
    else:
      print(f'Withdraw successful! New balance is: {self.tak}')

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```
